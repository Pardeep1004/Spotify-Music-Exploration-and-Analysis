```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/Most_Streamed_Spotify_Songs_2024.csv", encoding="latin-1")
df.head()
```

```python
columns_to_drop = ['Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins','TikTok Posts',
       'TikTok Likes', 'TikTok Views','ISRC','Amazon Playlist Count', 'Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams','TIDAL Popularity',"Deezer Playlist Reach" ,"Deezer Playlist Count","YouTube Playlist Reach","YouTube Likes","YouTube Views" ]
df = df.drop(columns_to_drop, axis = 1 )
```

```python
def summarize_data(data):
  """
  Provides a summary of a pandas DataFrame, including:

  * Number of unique values per column
  * Number of missing values per column
  * Data types of each column
  * Percentage of missing values per column (rounded to 2 decimal places)

  Args:
      data (pandas.DataFrame): The DataFrame to summarize.

  Returns:
      pandas.DataFrame: A DataFrame containing the summary statistics.
  """

  summary_df = pd.DataFrame({
      "Unique Values": data.nunique(),
      "Missing Values": data.isna().sum(),
      "Data Type": data.dtypes,
      "% Missing Values": round((data.isna().sum() / len(data)) * 100, 2)
  })

  return summary_df
```

```python
summarize_data(df)
```

```python
#Define columns for processing
columns_to_convert = [
    "All Time Rank", "Spotify Streams", "Spotify Playlist Count",
    "Spotify Playlist Reach", "Shazam Counts"
]

# Remove commas and convert to numeric (handling errors)
for col in columns_to_convert:
  # Convert to string (if necessary)
  if not pd.api.types.is_string_dtype(df[col]):
    df[col] = df[col].astype(str)
  # Remove commas
  df[col] = df[col].str.replace(",", "", regex=False)
  # Convert to numeric (handling errors)
  try:
    df[col] = pd.to_numeric(df[col], errors='coerce')
  except pd.errors.ParserError:
    print(f"Warning: Error converting column '{col}' to numeric.")

# Convert Release Date and extract year
df["Release Date"] = pd.to_datetime(df["Release Date"])
df["Year"] = df["Release Date"].dt.year
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Group data by year and calculate relevant statistics
album_counts_per_year = (
    df.groupby("Year")
    .agg(
        Album_Count=("Album Name", "count"),  # Count unique albums per year
        Stream_Total=("Spotify Streams", "sum"),  # Sum total streams per year
    )
    .reset_index()
)

# Filter data for desired year range (adjust as needed)
filtered_data = album_counts_per_year[(album_counts_per_year["Year"] > 2005) & (album_counts_per_year["Year"] < 2024)]

# Create subplots figure
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Stream Count trace (secondary y-axis)
trace1 = go.Scatter(
    x=filtered_data["Year"],
    y=filtered_data["Stream_Total"],
    mode="lines",
    name="Number of Streams",
    line=dict(color="blue", width=3),
)
fig.add_trace(trace1, secondary_y=True)

# Album Count trace (primary y-axis)
trace2 = go.Scatter(
    x=filtered_data["Year"],
    y=filtered_data["Album_Count"],
    mode="lines",
    name="Number of Albums",
    line=dict(color="#1DB954", width=3),
)
fig.add_trace(trace2, secondary_y=False)

# Layout customization
fig.update_layout(
    width=1200,
    height=500,
    paper_bgcolor="#FFFFFF",
    plot_bgcolor="#FFFFFF",
    font=dict(color="#000000"),
    title_font=dict(color="#1DB954"),
    title="<b>Albums Added and Streams per Year on Spotify<b>",
    legend=dict(title="", orientation="h", x=0.5, xanchor="center", y=1.1, yanchor="top"),
    showlegend=True,
)

# Update axes labels
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Number of Albums", secondary_y=False)
fig.update_yaxes(title_text="Number of Streams", secondary_y=True)

fig.show()
```

```python
df["Explicit Track"] = df["Explicit Track"].astype(str).replace({"0": "Non-explicit content", "1": "Explicit content"})

# Calculate the number of unique albums and total streams by year and content type
album_stats_per_year = (
    df.groupby(["Year", "Explicit Track"])
    .agg(
        Unique_Albums=("Album Name", "nunique"),  # Count unique albums
        Total_Streams=("Spotify Streams", "sum"),  # Sum the streams
    )
    .reset_index()
    .rename(columns={"Total_Streams": "Stream Count"})  # Rename for clarity
)

# Convert the "Explicit Track" column to string for better compatibility with Plotly
album_stats_per_year["Explicit Track"] = album_stats_per_year["Explicit Track"].astype(str)

# Filter data for years after 2005
filtered_album_stats = album_stats_per_year[album_stats_per_year["Year"] > 2005]

fig = make_subplots(rows=1, cols=2, specs=[[{"type": "bar"}, {"type": "domain"}]])

# Create a bar chart for stream counts
bar_plot = px.bar(
    filtered_album_stats,
    x="Year",
    y="Stream Count",
    color="Explicit Track",
    barmode="group",
    color_discrete_sequence=["Blue", "#1DB954"],  # Black and Spotify green
)
for trace in bar_plot.data:
    trace.showlegend = False
    fig.add_trace(trace, row=1, col=1)

# Calculate proportions for each content type to determine the largest portion
total_streams_by_type = filtered_album_stats.groupby("Explicit Track")["Stream Count"].sum()
pull_factors = [0.1 if value == total_streams_by_type.max() else 0 for value in total_streams_by_type]

# Create a pie chart
pie_plot = px.pie(
    filtered_album_stats,
    values="Stream Count",
    names="Explicit Track"
)

# Update pie chart colors to match bar chart and add pull factors
pie_plot.update_traces(marker=dict(colors=["#Blue", "#1DB954"]), pull=pull_factors)

# Add the pie chart trace to the figure
fig.add_trace(pie_plot.data[0], row=1, col=2)

# Update layout for both charts
fig.update_layout(
    width=1500,
    height=500,
    paper_bgcolor="#FFFFFF",
    plot_bgcolor="#FFFFFF",
    font=dict(color="#000000"),
    legend=dict(title="Content Type"),
    showlegend=True,
    annotations=[
        dict(text="<b>Explicit vs Non-explicit Albums / Year<b>", x=0.0, y=1, showarrow=False, xref="paper", yref="paper", font=dict(size=14)),
        dict(text="<b>Total Streams by Content Type<b>", x=0.80, y=1, showarrow=False, xref="paper", yref="paper", font=dict(size=14))
    ], # Shift pie chart title to the right
)

# Update axes
fig.update_xaxes(title_text="Year", row=1, col=1)  # X-axis for bar chart
fig.update_yaxes(title_text="Stream Count", row=1, col=1)  # Y-axis for bar chart

# Resize the pie chart
fig.update_traces(domain=dict(x=[0.6, 0.80]), row=1, col=2)  # Adjust values for desired size

fig.show()
```

```python
# Group by explicit track and calculate mean track score and mean Spotify popularity
explicit_stats = df.groupby("Explicit Track").agg(
    Average_Track_Score=('Track Score', 'mean'),
    Average_Spotify_Popularity=('Spotify Popularity', 'mean')
).reset_index()

# Reshape the dataframe for better visualization
explicit_stats_long = explicit_stats.melt(id_vars='Explicit Track',
                                          value_vars=['Average_Track_Score', 'Average_Spotify_Popularity'],
                                          var_name='Metric', value_name='Average Value')

# Create a horizontal bar chart with Plotly Express
fig = px.bar(explicit_stats_long,
             x='Average Value',
             y='Metric',
             color='Explicit Track',
             orientation='h',  # Horizontal bars
             title='<b>Average Track Scores and Spotify Popularity by Explicit Track<b>',
             barmode='group',
             color_discrete_sequence=["blue", "#1DB954"])  # Black and Spotify green

# Customize the layout
fig.update_layout(
    width=800,
    height=400,
    paper_bgcolor="#FFFFFF",
    plot_bgcolor="#FFFFFF",
    font=dict(color="#000000"),
    title_font=dict(color="#000000"),
    legend=dict(title="", orientation='h', x=0.5, xanchor='center', y=1.1, yanchor='top'),  # Position the legend
    showlegend=True  # Enable the legend
)

# Display the figure
fig.show()
```
